```python
import os
# import openai
import sys

# Google drive can be mounted from the left bar
from langchain.document_loaders import PyPDFLoader, PyPDFDirectoryLoader
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader

from langchain.text_splitter import RecursiveCharacterTextSplitter, CharacterTextSplitter

loader = PyPDFLoader("/content/drive/MyDrive/onedrive/health/Janaki/F042301093.pdf")
pages = loader.load_and_split(text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0))
#loader = DirectoryLoader("/content/drive/MyDrive/onedrive/health/Janaki", glob="**/*.pdf", loader_cls=TextLoader)
#docs = loader.load()
print(len(pages))
# print(pages[0])
# print(pages[1])
```

```python
from langchain.vectorstores import Chroma
#from langchain.embeddings import GooglePalmEmbeddings
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")

persist_directory = '/content/chroma/'
!rm -rf ./docs/chroma  # remove old database files if any

vectorstore = Chroma.from_documents(documents=pages, embedding=embeddings, persist_directory=persist_directory)
print(vectorstore._collection.count())
```

```python
question = "what is my hba1c"
docs = vectorstore.similarity_search(question,k=3)
len(docs)
# print(docs[0].page_content)
```

```python
question = "what is my hba1c"
docs = vectorstore.max_marginal_relevance_search(question,k=2, fetch_k=3)
# print(docs[0].page_content)
# print(docs[1].page_content)
```

```python
from litellm import completion
from typing import List, Dict
import json
import os
import pdb;
import litellm;

def generate_response(messages: List[Dict], tools) -> str:
    """Call LLM to get response"""
    response = completion(
        model="gemini/gemini-2.0-flash",
        messages=messages,
        tools=tools,
        max_tokens=1024
    )
    return response.choices[0].message.content


os.environ["GEMINI_API_KEY"] = "AIzaSyDadeXRQLSThV67AEDQYDKSjYCcms9naDY"

# assert litellm.supports_web_search(model="gemini/gemini-2.0-flash") == True
litellm.set_verbose(True)

messages = [
    {"role": "system", "content": "You are an expert at analyzing lab reports"}
]

messages.append({
    "role": "user",
    "content": "You can use the following context."
})

for doc in docs[0:]:
  messages.append({
    "role": "user",
    "content": doc.page_content
  })

tools=[{
        "type": "web_search_preview",
        # "search_context_size": "low"  # Options: "low", "medium" (default), "high"
    }]

question = ""
while question != "exit":
  question = input("What would you like me to do? ")
  # adds chat history
  messages.append({"role": "user", "content": question})
  response = generate_response(messages, tools=tools)
  messages.append({"role": "assistant", "content": response})
  print("-------------------------")
  print(response)
```
