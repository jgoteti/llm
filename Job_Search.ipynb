```python
from litellm import completion
from typing import List, Dict
import json
import os
import pdb;
import litellm;

def generate_response(messages: List[Dict], tools) -> str:
    """Call LLM to get response"""
    response = completion(
        model="gemini/gemini-2.0-flash",
        messages=messages,
        tools=tools,
        max_tokens=1024
    )
    return response.choices[0].message.content

```

```python
messages = [
    {"role": "system", "content": "You are an expert at providing a list of websites for job search."}
]

question = "Give a list of the best websites for job postings in Generative AI. Each website should be listed in the form www.linkedin.com"
messages.append({"role": "user", "content": question})
response = generate_response(messages, tools=[])
print(response)
```
